```python
social_media['age_group']=0
social_media.loc[social_media['age']<=2, 'Age_group']="babies"
social_media.loc[(social_media['age']>2) & (social_media['age']<=12), 'age_group']="childrens"
social_media.loc[(social_media['age']>12) & (social_media['age']<=18), 'age_group']="teens"
social_media.loc[(social_media['age']>18)& (social_media['age']<=30), 'age_group']="adults"
social_media.loc[(social_media['age']>30)& (social_media['age']<=50), 'age_group']="middle age"
social_media.loc[(social_media['age']>50)& (social_media['age']<=70), 'age_group']="aged"
social_media.head()
```

```python
social_media.drop('Age_group',axis = 1,inplace = False)
```

**Data Visualization**

```python
plt =social_media.time_spent.value_counts().plot(kind='bar',color= ['b','g' ,'r' ,'c'],title = 'Time spent on different platforms')
plt.set_xlabel('time_spent')
plt.set_ylabel('platform')
```

```python
plt = social_media.age_group.value_counts().plot(kind='bar',color= ['m','y' ,'k' ,'r'],title='Time spent on different age groups')
plt.set_xlabel('age_group')
plt.set_ylabel('time_spent')
```

```python
plt = social_media.gender.value_counts().plot(kind='bar',color= ['tab:red','tab:purple','tab:brown'],title='Time spent on socialmedia by Gender')
plt.set_xlabel('gender')
plt.set_ylabel('time_spent')
```
